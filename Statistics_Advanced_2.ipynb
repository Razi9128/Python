Question 1: What is hypothesis testing in statistics?

Hypothesis testing is a statistical method used to make inferences or decisions about a population based on sample data. It helps determine whether a specific claim or assumption about a population parameter is likely to be true.

Here’s how it works:

1. **Formulate Hypotheses**:
   - **Null Hypothesis (H₀)**: Assumes no effect or no difference. It's the default or status quo.
   - **Alternative Hypothesis (H₁ or Ha)**: Represents the claim you're testing—usually that there is an effect or a difference.

2. **Choose a Significance Level (α)**:
   - Common choices are 0.05, 0.01, or 0.10. This represents the probability of rejecting H₀ when it is actually true (Type I error).

3. **Select a Test and Compute the Test Statistic**:
   - Depending on the data and hypothesis, you might use a z-test, t-test, chi-square test, etc.

4. **Calculate the p-value**:
   - The p-value tells you how likely your sample results are, assuming H₀ is true.

5. **Make a Decision**:
   - If p-value ≤ α: Reject the null hypothesis (evidence supports H₁).
   - If p-value > α: Fail to reject the null hypothesis (insufficient evidence to support H₁).



Question 2: What is the null hypothesis, and how does it differ from the alternative
hypothesis?

The **null hypothesis (H₀)** is a statement that assumes there is no effect, no difference, or no relationship between variables. It serves as the default or starting assumption in hypothesis testing. Researchers test this hypothesis to determine whether it can be rejected based on sample data.

The **alternative hypothesis (H₁ or Ha)** is the statement that contradicts the null hypothesis. It suggests that there is an effect, a difference, or a relationship. This is the hypothesis that researchers typically hope to support.

### Key Differences:
- **H₀ (Null Hypothesis)**: Assumes no change or effect. Example: “The average score before and after training is the same.”
- **H₁ (Alternative Hypothesis)**: Assumes a change or effect exists. Example: “The average score after training is higher than before.”



Question 3: Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.


The **significance level**, often denoted by **α (alpha)**, is a threshold used in hypothesis testing to determine whether to reject the null hypothesis. It represents the **maximum probability of making a Type I error**, which occurs when the null hypothesis is rejected even though it is actually true.

---

### Purpose of Significance Level

- It sets the **standard for evidence** required to reject the null hypothesis.
- Common values are **0.05**, **0.01**, or **0.10**, depending on how strict the test needs to be.
  - **α = 0.05** means you're willing to accept a 5% chance of wrongly rejecting H₀.

---

### Role in Decision-Making

1. **Compare p-value to α**:
   - If **p-value ≤ α**: Reject the null hypothesis (evidence is strong).
   - If **p-value > α**: Fail to reject the null hypothesis (evidence is weak).

2. **Controls Risk**:
   - A lower α (e.g., 0.01) reduces the risk of false positives but increases the chance of missing a true effect (Type II error).
   - A higher α (e.g., 0.10) makes it easier to detect effects but increases the risk of false positives.

---

### Example

Suppose you're testing whether a new RPA training method improves scores:
- You set **α = 0.05**.
- After analyzing the data, the **p-value = 0.03**.
- Since **0.03 < 0.05**, you reject H₀ and conclude the method likely improves scores.

The significance level acts like a gatekeeper—it helps you decide when the evidence is strong enough to challenge the default assumption.

Question 4: What are Type I and Type II errors? Give examples of each.

In hypothesis testing, there are two types of errors that can occur when making decisions based on sample data:

**Type I Error (False Positive)**  
This error occurs when the null hypothesis (H₀) is rejected even though it is actually true.  
- It means you conclude there is an effect or difference when none exists.  
- The probability of making a Type I error is denoted by the significance level α (commonly 0.05).  
**Example**: A company tests a new automation tool and concludes it improves efficiency, but in reality, it doesn’t. The test wrongly rejects H₀.

**Type II Error (False Negative)**  
This error occurs when the null hypothesis is not rejected even though the alternative hypothesis (H₁) is actually true.  
- It means you fail to detect a real effect or difference.  
- The probability of making a Type II error is denoted by β.  
**Example**: A researcher evaluates a new SQL optimization technique that truly speeds up queries, but the test fails to show a significant improvement. The test wrongly retains H₀.

In summary:
- **Type I Error**: Rejecting a true null hypothesis.
- **Type II Error**: Failing to reject a false null hypothesis.

Balancing these errors is crucial in designing reliable statistical tests.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use
each.

A **Z-test** and a **T-test** are both statistical methods used to compare sample data to a population or between groups, but they differ in their assumptions and use cases.

---

### Z-test

**Use when**:
- The population standard deviation is known.
- The sample size is large (typically **n ≥ 30**).

**Purpose**:
- To test hypotheses about population means or proportions.

**Example**:
You want to test whether the average time to complete an RPA task is 10 minutes, and you know the population standard deviation. A Z-test is appropriate.

---

### T-test

**Use when**:
- The population standard deviation is **unknown**.
- The sample size is **small** (typically **n < 30**).

**Purpose**:
- To test hypotheses about population means when variability is estimated from the sample.

**Types**:
- **One-sample T-test**: Compare sample mean to a known value.
- **Two-sample T-test**: Compare means of two independent groups.
- **Paired T-test**: Compare means from the same group at two different times.

**Example**:
You’re testing whether a new SQL training module improves scores, but you only have a small sample and don’t know the population standard deviation. A T-test is the right choice.

---

### Summary of Differences

| Feature                     | Z-test                              | T-test                              |
|-----------------------------|--------------------------------------|-------------------------------------|
| Population standard deviation | Known                               | Unknown                             |
| Sample size                 | Large (n ≥ 30)                      | Small (n < 30)                      |
| Distribution used           | Normal distribution                 | t-distribution                      |
| Application                 | Proportions, large-sample means     | Small-sample means                  |

Choosing the right test ensures accurate conclusions. If you're building tutorials for your channel, this is a great concept to visualize with side-by-side examples.

```python
#Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
#Python. Show the Python code and interpret the results.
#sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
#                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
#                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
#                        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
#(Include your Python code and output in the code box below.)


import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
mu_0 = 50.0

# Significance level
alpha = 0.05

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z-statistic
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Output
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Standard Deviation: {sample_std:.4f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from 50.0")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from 50.0")
```

```python
#Question 8: Write a Python script to simulate data from a normal distribution and
#calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
#(Include your Python code and output in the code box below.)

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Set Seaborn style
sns.set(style="whitegrid")

# Parameters for the normal distribution
mean = 50
std_dev = 10
sample_size = 1000

# Simulate data
data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

# Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
sem = sample_std / np.sqrt(sample_size)

# Compute 95% confidence interval
confidence_level = 0.95
margin = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1) * sem
confidence_interval = (sample_mean - margin, sample_mean + margin)

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")

# Plot the data
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=30, kde=True, color='skyblue')
plt.axvline(sample_mean, color='blue', linestyle='-', label=f"Mean = {sample_mean:.2f}")
plt.axvline(confidence_interval[0], color='red', linestyle='--', label=f"Lower CI = {confidence_interval[0]:.2f}")
plt.axvline(confidence_interval[1], color='green', linestyle='--', label=f"Upper CI = {confidence_interval[1]:.2f}")
plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Question 9: Write a Python function to calculate the Z-scores from a dataset and
#visualize the standardized data using a histogram. Explain what the Z-scores represent
##in terms of standard deviations from the mean.
#(Include your Python code and output in the code box below.)

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set seaborn style for better aesthetics
sns.set(style="whitegrid")

# Generate a sample dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)  # Normal distribution with mean=50, std=10

# Convert to DataFrame
df = pd.DataFrame(data, columns=['Value'])

# Calculate Z-scores
mean = df['Value'].mean()
std = df['Value'].std()
df['Z-Score'] = (df['Value'] - mean) / std

# Plot histogram of Z-scores
plt.figure(figsize=(10, 6))
sns.histplot(df['Z-Score'], bins=30, kde=True, color='skyblue')
plt.title('Histogram of Z-Scores')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
plt.legend()
plt.tight_layout()
plt.show()

# Print summary statistics
print("Mean of original data:", round(mean, 2))
print("Standard deviation of original data:", round(std, 2))
print("First 5 Z-scores:\n", df['Z-Score'].head())

```
